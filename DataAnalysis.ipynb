```python
%matplotlib inline
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
import csv
```

```python
df1=pd.read_csv("/Users/nikhila/Desktop/Courses/second_sem/256/Project/Code/data/preprocessed_data.csv", low_memory=False)
```

```python
nh = Counter(df1['bedroomcnt'])
nh
```

```python
nh_df = pd.DataFrame.from_dict(nh, orient='index').sort_values(by=0)
nh_df.columns = ['bedroomcnt']
nh_df.plot.bar(y= 'bedroomcnt', 
                 color = 'LightBlue',
                 fontsize = 20,
                 legend = False,
                 figsize= (20, 10),
                 title = "Number houses distribution based on BedRoomCount")
```

```python
r = df1.poolcnt
nh = Counter(r)
nh
```

```python
nh_df = pd.DataFrame.from_dict(nh, orient='index').sort_values(by=0)
nh_df.columns =['poolcnt']
nh_df.plot.pie(y = 'poolcnt', 
                 colormap = 'Greens_r', 
                 figsize=(6,6), 
                 fontsize = 18, autopct = '%.2f',
                 legend = False,
                 title = 'House with Pool Count Distribution')
```
