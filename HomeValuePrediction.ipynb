```python
df2 = df1.loc[:,['parcelid','lotsizesquarefeet', 'calculatedfinishedsquarefeet', 'landtaxvaluedollarcnt','structuretaxvaluedollarcnt','bedroomcnt','bathroomcnt','yearbuilt','regionidzip','garagecarcnt','poolcnt','buildingclasstypeid','taxvaluedollarcnt']]
```

```python
df2['poolcnt']=df2['poolcnt'].fillna(0)
df2['buildingclasstypeid']=df2['buildingclasstypeid'].fillna(0)
df2['garagecarcnt']=df2['garagecarcnt'].fillna(0)
df2['taxvaluedollarcnt']=df2['taxvaluedollarcnt'].fillna(0)
df2.head()
```

```python
df3=df2.loc[(df2['taxvaluedollarcnt'] > 900000) & (df2['taxvaluedollarcnt'] < 1300000)]
df3.count()
```

```python
df3=df3[df3['bedroomcnt']!=0 & (df3['bathroomcnt'] != 0) ]
df3['AgeofHouse']=2017-df3['yearbuilt']
df3['lotpricepersqft']=df3['landtaxvaluedollarcnt']/df3['lotsizesquarefeet']
df3['buildingpricepersqft']=df3['structuretaxvaluedollarcnt']/df3['calculatedfinishedsquarefeet']
df3.regionidzip = pd.factorize(df3.regionidzip)[0]
```

```python
df4 = df3.loc[:,['lotsizesquarefeet', 'calculatedfinishedsquarefeet','bedroomcnt','bathroomcnt','regionidzip','garagecarcnt','poolcnt','AgeofHouse','buildingclasstypeid']]
df4.count()
```

```python
def min_max(series):
    mean = series.min()
    series_min = series.max()
    series_max = series.min()
    diff1 = series-mean
    diff2 = series_max - series_min    
    return diff1/diff2 
```

```python
df4['lotsizesquarefeet'] = min_max(df3['lotsizesquarefeet'])
df4.head(5)

df4['calculatedfinishedsquarefeet'] = min_max(df3['lotsizesquarefeet'])
df4.head(5)


df4.head(5)
df4.count()
```
